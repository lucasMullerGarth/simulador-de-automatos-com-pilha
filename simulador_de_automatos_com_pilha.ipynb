## Implement transition logic

### Subtask:
Develop functions to handle transitions based on the current state, input symbol, and top of the stack.


**Reasoning**:
Implement the `get_transitions` and `apply_transition` methods within the `PushdownAutomaton` class as described in the instructions.

### Subtask:
Create a mechanism to process the input string one symbol at a time, updating the state, stack, and remaining input at each step.


**Reasoning**:
Add the `simulate_step` and `process_input` methods to the `PushdownAutomaton` class as described in the instructions. `simulate_step` will handle processing a single input symbol (or epsilon), and `process_input` will iterate through the input string, calling `simulate_step` and managing the remaining input. I will also add a way to store the remaining input string within the `PushdownAutomaton` class and update it during the simulation.



```python
import tkinter as tk
from tkinter import ttk, messagebox

class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if not self.is_empty():
            return self._items.pop()
        return None

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        return None

    def is_empty(self):
        return not self._items

    def __str__(self):
        return str(self._items)

    def __repr__(self):
        return repr(self._items)

class PushdownAutomaton:
    def __init__(self, states, input_alphabet, stack_alphabet, initial_state, final_states, transitions):
        self.states = set(states)
        self.input_alphabet = set(input_alphabet)
        self.stack_alphabet = set(stack_alphabet)
        self.initial_state = initial_state
        self.final_states = set(final_states)
        self.transitions = transitions

        self.current_state = initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0 # Adicionado para rastrear a posição atual na string de entrada

    def reset(self):
        self.current_state = self.initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0

    def get_transitions(self, current_state, input_symbol, stack_top):
        """
        Procura transições possíveis com base no estado atual, símbolo de entrada e topo da pilha.

        Args:
            current_state: O estado atual do autômato.
            input_symbol: O símbolo de entrada atual sendo lido (ou string vazia para epsilon).
            stack_top: O símbolo no topo da pilha (ou None se a pilha estiver vazia).

        Returns:
            A lista de tuplas possíveis (próximo_estado, string_para_empilhar).
            Retorna uma lista vazia se nenhuma transição correspondente for encontrada.
        """ # Traduzido
        key = (current_state, input_symbol, stack_top)
        return self.transitions.get(key, [])

    def apply_transition(self, transition):
        """
        Aplica uma transição dada para atualizar o estado e a pilha do autômato.

        Args:
            transition: Uma tupla (próximo_estado, string_para_empilhar).
        """ # Traduzido
        next_state, stack_push_string = transition
        self.current_state = next_state

        # Remover o topo da pilha se a regra de transição exigiu a leitura de um símbolo da pilha
        # (assumindo que get_transitions foi chamado com um stack_top não vazio, se aplicável)
        if self.stack.peek() is not None:
             self.stack.pop()


        # Empilhar os caracteres da string_para_empilhar na pilha em ordem inversa
        for char in reversed(stack_push_string):
            self.stack.push(char)

    def simulate_step(self, input_symbol):
        """
        Processa um único símbolo de entrada (ou epsilon) e atualiza o estado e a pilha do autômato.

        Args:
            input_symbol: O símbolo de entrada atual a ser processado (ou string vazia para epsilon).
        """ # Traduzido
        if self.is_halted:
            return

        current_state = self.current_state
        stack_top = self.stack.peek()

        # Tentar transições com o símbolo de entrada real
        possible_transitions = self.get_transitions(current_state, input_symbol, stack_top)

        # Se não houver transições com o símbolo de entrada, tentar transições epsilon
        if not possible_transitions:
            possible_transitions = self.get_transitions(current_state, '', stack_top)
            input_consumed = False # Transição epsilon, entrada não consumida
        else:
            input_consumed = True # Transição não epsilon, entrada consumida


        if possible_transitions:
            # Para simplificar, lidar com não-determinismo escolhendo a primeira transição válida
            chosen_transition = possible_transitions[0]
            self.apply_transition(chosen_transition)
            return input_consumed # Indicar se a entrada foi consumida
        else:
            # Nenhuma transição válida encontrada, parar a simulação
            self.is_halted = True
            return False # A entrada não foi consumida se a simulação parou


    def process_input(self, input_string):
        """
        Itera pela string de entrada, simulando o AP passo a passo.

        Args:
            input_string: A string a ser processada.
        """ # Traduzido
        self.reset()
        self.remaining_input = input_string

        while self.remaining_input or self.get_transitions(self.current_state, '', self.stack.peek()):
             if self.is_halted:
                 break

             # Priorizar transições epsilon se disponíveis
             epsilon_transitions = self.get_transitions(self.current_state, '', self.stack.peek())
             if epsilon_transitions:
                 # Aplicar a primeira transição epsilon encontrada
                 self.apply_transition(epsilon_transitions[0])
                 # Não consumir entrada
             elif self.remaining_input:
                 # Processar o próximo símbolo de entrada
                 current_symbol = self.remaining_input[0]
                 input_consumed = self.simulate_step(current_symbol)
                 if input_consumed:
                     self.remaining_input = self.remaining_input[1:]
             else:
                 # Não há mais entrada e nem transições epsilon
                 self.is_halted = True


    def is_accepted(self):
        """
        Verifica se a string de entrada é aceita pelo AP.
        A aceitação pode ser por estado final, pilha vazia ou ambos, dependendo da definição do AP.
        Para esta implementação, verificaremos a aceitação por estado final E pilha vazia
        após processar a string de entrada inteira.
        """ # Traduzido
        return not self.remaining_input and not self.stack.is_empty() and self.current_state in self.final_states


# A classe PDASimulatorGUI permanece a mesma do passo anterior,
# mas precisará interagir com os novos métodos em PushdownAutomaton
class PDASimulatorGUI:
    def __init__(self, master):
        self.master = master
        master.title("Simulador de Autômato com Pilha") # Traduzido

        # Quadro de Entrada
        self.input_frame = ttk.LabelFrame(master, text="Definição do AP") # Traduzido
        self.input_frame.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")

        self.alphabet_label = ttk.Label(self.input_frame, text="Alfabeto (separado por vírgulas):") # Traduzido
        self.alphabet_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.alphabet_entry = ttk.Entry(self.input_frame)
        self.alphabet_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.states_label = ttk.Label(self.input_frame, text="Estados (separados por vírgulas):") # Traduzido
        self.states_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.states_entry = ttk.Entry(self.input_frame)
        self.states_entry.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.initial_state_label = ttk.Label(self.input_frame, text="Estado Inicial:") # Traduzido
        self.initial_state_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.initial_state_entry = ttk.Entry(self.input_frame)
        self.initial_state_entry.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.final_states_label = ttk.Label(self.input_frame, text="Estados Finais (separados por vírgulas):") # Traduzido
        self.final_states_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.final_states_entry = ttk.Entry(self.input_frame)
        self.final_states_entry.grid(row=3, column=1, padx=5, pady=5, sticky="ew")

        self.stack_alphabet_label = ttk.Label(self.input_frame, text="Alfabeto da Pilha (separado por vírgulas):") # Traduzido
        self.stack_alphabet_label.grid(row=4, column=0, padx=5, pady=5, sticky="w")
        self.stack_alphabet_entry = ttk.Entry(self.input_frame)
        self.stack_alphabet_entry.grid(row=4, column=1, padx=5, pady=5, sticky="ew")

        self.transitions_label = ttk.Label(self.input_frame, text="Transições (uma por linha, formato: q,a,S->p,StringParaEmpilhar):") # Traduzido
        self.transitions_label.grid(row=5, column=0, padx=5, pady=5, sticky="nw")
        self.transitions_text = tk.Text(self.input_frame, height=10, width=50)
        self.transitions_text.grid(row=5, column=1, padx=5, pady=5, sticky="nsew")

        self.process_transitions_button = ttk.Button(self.input_frame, text="Processar Transições", command=self.process_transitions) # Traduzido
        self.process_transitions_button.grid(row=6, column=0, columnspan=2, padx=5, pady=5)


        # Quadro de Entrada de String
        self.input_string_frame = ttk.LabelFrame(master, text="Entrada para Simulação") # Traduzido
        self.input_string_frame.grid(row=0, column=1, padx=10, pady=10, sticky="nsew")

        self.input_string_label = ttk.Label(self.input_string_frame, text="String de Entrada:") # Traduzido
        self.input_string_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.input_string_entry = ttk.Entry(self.input_string_frame)
        self.input_string_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.start_simulation_button = ttk.Button(self.input_string_frame, text="Iniciar Simulação", command=self.start_simulation) # Traduzido
        self.start_simulation_button.grid(row=1, column=0, columnspan=2, padx=5, pady=5)


        # Quadro de Saída
        self.output_frame = ttk.LabelFrame(master, text="Saída da Simulação") # Traduzido
        self.output_frame.grid(row=1, column=0, columnspan=2, padx=10, pady=10, sticky="nsew")

        self.current_state_label = ttk.Label(self.output_frame, text="Estado Atual:") # Traduzido
        self.current_state_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.current_state_display = ttk.Label(self.output_frame, text="")
        self.current_state_display.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.stack_content_label = ttk.Label(self.output_frame, text="Conteúdo da Pilha:") # Traduzido
        self.stack_content_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.stack_content_display = ttk.Label(self.output_frame, text="")
        self.stack_content_display.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.remaining_input_label = ttk.Label(self.output_frame, text="Entrada Restante:") # Traduzido
        self.remaining_input_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.remaining_input_display = ttk.Label(self.output_frame, text="")
        self.remaining_input_display.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.acceptance_status_label = ttk.Label(self.output_frame, text="Status de Aceitação:") # Traduzido
        self.acceptance_status_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.acceptance_status_display = ttk.Label(self.output_frame, text="")
        self.acceptance_status_display.grid(row=3, column=1, padx=5, pady=5, sticky="ew")

        self.pda = None
        self.parsed_transitions = {}

    def process_transitions(self):
        transitions_input = self.transitions_text.get("1.0", "end-1c")
        transitions = {}
        errors = []

        for line_num, line in enumerate(transitions_input.splitlines()):
            line = line.strip()
            if not line:
                continue

            try:
                # Formato esperado: q,a,S->p,StringParaEmpilhar
                parts = line.split("->")
                if len(parts) != 2:
                    errors.append(f"Linha {line_num+1}: Formato inválido. Esperado 'q,a,S->p,StringParaEmpilhar'.") # Traduzido
                    continue

                left_part = parts[0].split(",")
                right_part = parts[1].split(",")

                if len(left_part) != 3 or len(right_part) != 2:
                     errors.append(f"Linha {line_num+1}: Número inválido de elementos na regra de transição.") # Traduzido
                     continue

                current_state, input_symbol, stack_top = left_part
                next_state, stack_push_string = right_part

                # Usar 'epsilon' ou uma convenção similar para transições epsilon
                if input_symbol.lower() == 'epsilon':
                    input_symbol = ''
                if stack_top.lower() == 'epsilon':
                    stack_top = ''
                if stack_push_string.lower() == 'epsilon':
                    stack_push_string = ''


                # Armazenar transições em um dicionário: (estado_atual, símbolo_entrada, topo_pilha) -> lista de (próximo_estado, string_para_empilhar)
                key = (current_state.strip(), input_symbol.strip(), stack_top.strip())
                value = (next_state.strip(), stack_push_string.strip())

                if key not in transitions:
                    transitions[key] = []
                transitions[key].append(value)

            except Exception as e:
                errors.append(f"Linha {line_num+1}: Erro ao analisar transição - {e}") # Traduzido


        if errors:
            messagebox.showerror("Erro na Entrada de Transição", "\n".join(errors)) # Traduzido
            self.parsed_transitions = {}
        else:
            messagebox.showinfo("Entrada de Transição", "Transições processadas com sucesso.") # Traduzido
            self.parsed_transitions = transitions
            print("Transições Analisadas:", self.parsed_transitions)

    def start_simulation(self):
        # Obter a definição do AP a partir dos campos de entrada da GUI
        states = [s.strip() for s in self.states_entry.get().split(',') if s.strip()]
        input_alphabet = [s.strip() for s in self.alphabet_entry.get().split(',') if s.strip()]
        stack_alphabet = [s.strip() for s in self.stack_alphabet_entry.get().split(',') if s.strip()]
        initial_state = self.initial_state_entry.get().strip()
        final_states = [s.strip() for s in self.final_states_entry.get().split(',') if s.strip()]
        input_string = self.input_string_entry.get().strip()

        if not self.parsed_transitions:
            messagebox.showwarning("Erro de Simulação", "Por favor, processe as transições primeiro.") # Traduzido
            return

        # Criar a instância do AP
        try:
            self.pda = PushdownAutomaton(
                states=states,
                input_alphabet=input_alphabet,
                stack_alphabet=stack_alphabet,
                initial_state=initial_state,
                final_states=final_states,
                transitions=self.parsed_transitions
            )
            self.pda.remaining_input = input_string
            self.pda.input_index = 0
            self.pda.reset()

            self.update_gui_output()
            self.acceptance_status_display.config(text="")
            self.next_step_button.config(state=tk.NORMAL)
            self.start_simulation_button.config(state=tk.DISABLED)

        except Exception as e:
            messagebox.showerror("Erro na Criação do AP", f"Erro ao criar AP: {e}") # Traduzido


    def update_gui_output(self):
        if self.pda:
            self.current_state_display.config(text=self.pda.current_state)
            self.stack_content_display.config(text="".join(self.pda.stack._items))
            self.remaining_input_display.config(text=self.pda.remaining_input[self.pda.input_index:])
        else:
             self.current_state_display.config(text="N/A")
             self.stack_content_display.config(text="N/A")
             self.remaining_input_display.config(text="N/A")


    def check_acceptance(self):
        if self.pda and self.pda.is_halted:
            self.acceptance_status_display.config(text="Parado (Rejeitado)", foreground="red") # Traduzido
        elif self.pda and self.pda.is_accepted():
            self.acceptance_status_display.config(text="Aceito", foreground="green") # Traduzido
        elif self.pda:
             self.acceptance_status_display.config(text="Rejeitado", foreground="red") # Traduzido
        else:
             self.acceptance_status_display.config(text="N/A")


# Criar a janela principal
# root = tk.Tk()
# pda_gui = PDASimulatorGUI(root)
# root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox

class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if not self.is_empty():
            return self._items.pop()
        return None  # Or raise an error for an empty stack

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        return None

    def is_empty(self):
        return not self._items

    def __str__(self):
        return str(self._items)

    def __repr__(self):
        return repr(self._items)

class PushdownAutomaton:
    def __init__(self, states, input_alphabet, stack_alphabet, initial_state, final_states, transitions):
        self.states = set(states)
        self.input_alphabet = set(input_alphabet)
        self.stack_alphabet = set(stack_alphabet)
        self.initial_state = initial_state
        self.final_states = set(final_states)
        self.transitions = transitions  # Using a dictionary for transitions

        self.current_state = initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0 # Added to track current position in input string

    def reset(self):
        self.current_state = self.initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0

    def get_transitions(self, current_state, input_symbol, stack_top):
        """
        Looks up possible transitions based on current state, input symbol, and stack top.

        Args:
            current_state: The current state of the automaton.
            input_symbol: The current input symbol being read (or empty string for epsilon).
            stack_top: The symbol at the top of the stack (or None if stack is empty).

        Returns:
            A list of possible (next_state, stack_push_string) tuples.
            Returns an empty list if no matching transitions are found.
        """
        key = (current_state, input_symbol, stack_top)
        return self.transitions.get(key, [])

    def apply_transition(self, transition):
        """
        Applies a given transition to update the automaton's state and stack.

        Args:
            transition: A tuple (next_state, stack_push_string).
        """
        next_state, stack_push_string = transition
        self.current_state = next_state

        # Pop the top of the stack if the transition rule required reading a stack symbol
        # (assuming get_transitions was called with a non-empty stack_top if applicable)
        if self.stack.peek() is not None: # Only pop if the stack is not empty
             self.stack.pop()


        # Push the stack_push_string characters onto the stack in reverse order
        for char in reversed(stack_push_string):
            self.stack.push(char)

    def step_simulation(self):
        """
        Performs a single step of the PDA simulation.

        Returns:
            A status string: "continue", "halted", or "accepted".
        """
        if self.is_halted:
            return "halted"

        current_state = self.current_state
        stack_top = self.stack.peek()
        current_symbol = self.remaining_input[self.input_index] if self.input_index < len(self.remaining_input) else ''

        # Try transitions with the actual input symbol
        possible_transitions = self.get_transitions(current_state, current_symbol, stack_top)

        # If no transitions with the input symbol, try epsilon transitions
        if not possible_transitions:
            possible_transitions = self.get_transitions(current_state, '', stack_top)
            input_consumed = False # Epsilon transition, input not consumed
        else:
            input_consumed = True # Non-epsilon transition, input is consumed


        if possible_transitions:
            # For simplicity, handle non-determinism by choosing the first valid transition
            chosen_transition = possible_transitions[0]
            self.apply_transition(chosen_transition)
            if input_consumed:
                self.input_index += 1 # Move to the next input symbol

            # Check if simulation is complete after this step
            if self.input_index >= len(self.remaining_input) and not self.get_transitions(self.current_state, '', self.stack.peek()):
                self.is_halted = True
                if self.current_state in self.final_states: # Acceptance by final state
                     return "accepted"
                else:
                     return "halted"
            else:
                return "continue"

        else:
            # No valid transition found, halt simulation
            self.is_halted = True
            return "halted"


    def is_accepted(self):
        """
        Checks if the input string is accepted by the PDA.
        Acceptance is by final state and empty stack after processing the entire input string.
        """
        return self.input_index >= len(self.remaining_input) and self.stack.is_empty() and self.current_state in self.final_states


class PDASimulatorGUI:
    def __init__(self, master):
        self.master = master
        master.title("Pushdown Automaton Simulator")

        # Input Frame
        self.input_frame = ttk.LabelFrame(master, text="PDA Definition")
        self.input_frame.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")

        self.alphabet_label = ttk.Label(self.input_frame, text="Alphabet (comma-separated):")
        self.alphabet_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.alphabet_entry = ttk.Entry(self.input_frame)
        self.alphabet_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.states_label = ttk.Label(self.input_frame, text="States (comma-separated):")
        self.states_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.states_entry = ttk.Entry(self.input_frame)
        self.states_entry.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.initial_state_label = ttk.Label(self.input_frame, text="Initial State:")
        self.initial_state_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.initial_state_entry = ttk.Entry(self.input_frame)
        self.initial_state_entry.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.final_states_label = ttk.Label(self.input_frame, text="Final States (comma-separated):")
        self.final_states_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.final_states_entry = ttk.Entry(self.input_frame)
        self.final_states_entry.grid(row=3, column=1, padx=5, pady=5, sticky="ew")

        self.stack_alphabet_label = ttk.Label(self.input_frame, text="Stack Alphabet (comma-separated):")
        self.stack_alphabet_label.grid(row=4, column=0, padx=5, pady=5, sticky="w")
        self.stack_alphabet_entry = ttk.Entry(self.input_frame)
        self.stack_alphabet_entry.grid(row=4, column=1, padx=5, pady=5, sticky="ew")

        self.transitions_label = ttk.Label(self.input_frame, text="Transitions (one per line, format: q,a,S->p,PushString):")
        self.transitions_label.grid(row=5, column=0, padx=5, pady=5, sticky="nw")
        self.transitions_text = tk.Text(self.input_frame, height=10, width=50)
        self.transitions_text.grid(row=5, column=1, padx=5, pady=5, sticky="nsew")

        self.process_transitions_button = ttk.Button(self.input_frame, text="Process Transitions", command=self.process_transitions)
        self.process_transitions_button.grid(row=6, column=0, columnspan=2, padx=5, pady=5)


        # Input String Frame
        self.input_string_frame = ttk.LabelFrame(master, text="Simulation Input")
        self.input_string_frame.grid(row=0, column=1, padx=10, pady=10, sticky="nsew")

        self.input_string_label = ttk.Label(self.input_string_frame, text="Input String:")
        self.input_string_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.input_string_entry = ttk.Entry(self.input_string_frame)
        self.input_string_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.start_simulation_button = ttk.Button(self.input_string_frame, text="Start Simulation", command=self.start_simulation)
        self.start_simulation_button.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

        self.next_step_button = ttk.Button(self.input_string_frame, text="Next Step", command=self.step_simulation_gui, state=tk.DISABLED)
        self.next_step_button.grid(row=2, column=0, columnspan=2, padx=5, pady=5)


        # Output Frame
        self.output_frame = ttk.LabelFrame(master, text="Simulation Output")
        self.output_frame.grid(row=1, column=0, columnspan=2, padx=10, pady=10, sticky="nsew")

        self.current_state_label = ttk.Label(self.output_frame, text="Current State:")
        self.current_state_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.current_state_display = ttk.Label(self.output_frame, text="")
        self.current_state_display.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.stack_content_label = ttk.Label(self.output_frame, text="Stack Content:")
        self.stack_content_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.stack_content_display = ttk.Label(self.output_frame, text="")
        self.stack_content_display.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.remaining_input_label = ttk.Label(self.output_frame, text="Remaining Input:")
        self.remaining_input_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.remaining_input_display = ttk.Label(self.output_frame, text="")
        self.remaining_input_display.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.acceptance_status_label = ttk.Label(self.output_frame, text="Acceptance Status:")
        self.acceptance_status_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.acceptance_status_display = ttk.Label(self.output_frame, text="")
        self.acceptance_status_display.grid(row=3, column=1, padx=5, pady=5, sticky="ew")

        self.pda = None # Placeholder for the PDA instance
        self.parsed_transitions = {} # To store transitions after processing

    def process_transitions(self):
        transitions_input = self.transitions_text.get("1.0", "end-1c")
        transitions = {}
        errors = []

        for line_num, line in enumerate(transitions_input.splitlines()):
            line = line.strip()
            if not line:
                continue

            try:
                # Expected format: q,a,S->p,PushString
                parts = line.split("->")
                if len(parts) != 2:
                    errors.append(f"Line {line_num+1}: Invalid format. Expected 'q,a,S->p,PushString'.")
                    continue

                left_part = parts[0].split(",")
                right_part = parts[1].split(",")

                if len(left_part) != 3 or len(right_part) != 2:
                     errors.append(f"Line {line_num+1}: Invalid number of elements in transition rule.")
                     continue

                current_state, input_symbol, stack_top = left_part
                next_state, stack_push_string = right_part

                # Use 'epsilon' or a similar convention for epsilon transitions
                if input_symbol.lower() == 'epsilon':
                    input_symbol = ''
                if stack_top.lower() == 'epsilon':
                    stack_top = ''
                if stack_push_string.lower() == 'epsilon':
                    stack_push_string = ''


                # Store transitions in a dictionary: (current_state, input_symbol, stack_top) -> list of (next_state, stack_push_string)
                key = (current_state.strip(), input_symbol.strip(), stack_top.strip())
                value = (next_state.strip(), stack_push_string.strip())

                if key not in transitions:
                    transitions[key] = []
                transitions[key].append(value)

            except Exception as e:
                errors.append(f"Line {line_num+1}: Error parsing transition - {e}")


        if errors:
            messagebox.showerror("Transition Input Error", "\n".join(errors))
            self.parsed_transitions = {} # Clear transitions if invalid
        else:
            messagebox.showinfo("Transition Input", "Transitions processed successfully.")
            self.parsed_transitions = transitions
            print("Parsed Transitions:", self.parsed_transitions) # For debugging

    def start_simulation(self):
        # Get PDA definition from GUI entries
        states = [s.strip() for s in self.states_entry.get().split(',') if s.strip()]
        input_alphabet = [s.strip() for s in self.alphabet_entry.get().split(',') if s.strip()]
        stack_alphabet = [s.strip() for s in self.stack_alphabet_entry.get().split(',') if s.strip()]
        initial_state = self.initial_state_entry.get().strip()
        final_states = [s.strip() for s in self.final_states_entry.get().split(',') if s.strip()]
        input_string = self.input_string_entry.get().strip()

        if not self.parsed_transitions:
            messagebox.showwarning("Simulation Error", "Please process transitions first.")
            return

        # Create the PDA instance
        try:
            self.pda = PushdownAutomaton(
                states=states,
                input_alphabet=input_alphabet,
                stack_alphabet=stack_alphabet,
                initial_state=initial_state,
                final_states=final_states,
                transitions=self.parsed_transitions
            )
            self.pda.remaining_input = input_string # Set the full input string
            self.pda.input_index = 0 # Reset input index
            self.pda.reset() # Reset PDA state and stack

            self.update_gui_output()
            self.acceptance_status_display.config(text="") # Clear previous status
            self.next_step_button.config(state=tk.NORMAL) # Enable Next Step button
            self.start_simulation_button.config(state=tk.DISABLED) # Disable Start Simulation button

        except Exception as e:
            messagebox.showerror("PDA Creation Error", f"Error creating PDA: {e}")


    def step_simulation_gui(self):
        if self.pda is None:
            messagebox.showwarning("Simulation Error", "Please start a simulation first.")
            return

        status = self.pda.step_simulation()
        self.update_gui_output()

        if status == "halted":
            self.acceptance_status_display.config(text="Halted (Rejected)", foreground="red")
            self.next_step_button.config(state=tk.DISABLED) # Disable Next Step button
            self.start_simulation_button.config(state=tk.NORMAL) # Enable Start Simulation button
        elif status == "accepted":
             self.acceptance_status_display.config(text="Accepted", foreground="green")
             self.next_step_button.config(state=tk.DISABLED) # Disable Next Step button
             self.start_simulation_button.config(state=tk.NORMAL) # Enable Start Simulation button
        elif status == "continue":
            # Simulation continues, nothing to do here
            pass


    def update_gui_output(self):
        if self.pda:
            self.current_state_display.config(text=self.pda.current_state)
            # Display stack content from bottom to top
            self.stack_content_display.config(text="".join(self.pda.stack._items))
            self.remaining_input_display.config(text=self.pda.remaining_input[self.pda.input_index:])
        else:
             self.current_state_display.config(text="N/A")
             self.stack_content_display.config(text="N/A")
             self.remaining_input_display.config(text="N/A")


    def check_acceptance(self):
        # This method is less relevant for step-by-step, acceptance is checked in step_simulation
        pass


# Create the main window
# root = tk.Tk()
# pda_gui = PDASimulatorGUI(root)
# root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox

class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if not self.is_empty():
            return self._items.pop()
        return None  # Or raise an error for an empty stack

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        return None

    def is_empty(self):
        return not self._items

    def __str__(self):
        return str(self._items)

    def __repr__(self):
        return repr(self._items)

class PushdownAutomaton:
    def __init__(self, states, input_alphabet, stack_alphabet, initial_state, final_states, transitions):
        self.states = set(states)
        self.input_alphabet = set(input_alphabet)
        self.stack_alphabet = set(stack_alphabet)
        self.initial_state = initial_state
        self.final_states = set(final_states)
        self.transitions = transitions  # Using a dictionary for transitions

        self.current_state = initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0 # Added to track current position in input string

        # Acceptance condition: 'final_state', 'empty_stack', or 'both'
        # Defaulting to 'both' as per previous steps, will be modifiable later
        self.acceptance_condition = 'both'

    def reset(self):
        self.current_state = self.initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0

    def get_transitions(self, current_state, input_symbol, stack_top):
        """
        Looks up possible transitions based on current state, input symbol, and stack top.

        Args:
            current_state: The current state of the automaton.
            input_symbol: The current input symbol being read (or empty string for epsilon).
            stack_top: The symbol at the top of the stack (or None if stack is empty).

        Returns:
            A list of possible (next_state, stack_push_string) tuples.
            Returns an empty list if no matching transitions are found.
        """
        key = (current_state, input_symbol, stack_top)
        return self.transitions.get(key, [])

    def apply_transition(self, transition):
        """
        Applies a given transition to update the automaton's state and stack.

        Args:
            transition: A tuple (next_state, stack_push_string).
        """
        next_state, stack_push_string = transition
        self.current_state = next_state

        # Pop the top of the stack if the transition rule required reading a stack symbol
        # (assuming get_transitions was called with a non-empty stack_top if applicable)
        if self.stack.peek() is not None: # Only pop if the stack is not empty
             self.stack.pop()


        # Push the stack_push_string characters onto the stack in reverse order
        for char in reversed(stack_push_string):
            self.stack.push(char)

    def step_simulation(self):
        """
        Performs a single step of the PDA simulation.

        Returns:
            A status string: "continue", "halted", or "accepted".
        """
        if self.is_halted:
            # Check for acceptance only if already halted
            if self.is_accepted():
                 return "accepted"
            else:
                 return "halted"


        current_state = self.current_state
        stack_top = self.stack.peek()
        current_symbol = self.remaining_input[self.input_index] if self.input_index < len(self.remaining_input) else ''

        # Try transitions with the actual input symbol
        possible_transitions = self.get_transitions(current_state, current_symbol, stack_top)

        # If no transitions with the input symbol, try epsilon transitions
        if not possible_transitions:
            possible_transitions = self.get_transitions(current_state, '', stack_top)
            input_consumed = False # Epsilon transition, input not consumed
        else:
            input_consumed = True # Non-epsilon transition, input is consumed


        if possible_transitions:
            # For simplicity, handle non-determinism by choosing the first valid transition
            chosen_transition = possible_transitions[0]
            self.apply_transition(chosen_transition)
            if input_consumed:
                self.input_index += 1 # Move to the next input symbol

            # After applying a transition, check if the simulation should halt
            # Simulation halts if no further transitions (input or epsilon) are possible
            next_symbol = self.remaining_input[self.input_index] if self.input_index < len(self.remaining_input) else ''
            if not self.get_transitions(self.current_state, next_symbol, self.stack.peek()) and \
               not self.get_transitions(self.current_state, '', self.stack.peek()):
                self.is_halted = True
                if self.is_accepted():
                    return "accepted"
                else:
                    return "halted"
            else:
                return "continue"

        else:
            # No valid transition found, halt simulation
            self.is_halted = True
            if self.is_accepted():
                 return "accepted"
            else:
                 return "halted"


    def is_accepted(self):
        """
        Checks if the input string is accepted by the PDA based on the defined acceptance condition.
        Current condition is final state AND empty stack after processing the entire input string.
        """
        # Acceptance by final state AND empty stack
        return self.input_index >= len(self.remaining_input) and self.stack.is_empty() and self.current_state in self.final_states


class PDASimulatorGUI:
    def __init__(self, master):
        self.master = master
        master.title("Pushdown Automaton Simulator")

        # Input Frame
        self.input_frame = ttk.LabelFrame(master, text="PDA Definition")
        self.input_frame.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")

        self.alphabet_label = ttk.Label(self.input_frame, text="Alphabet (comma-separated):")
        self.alphabet_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.alphabet_entry = ttk.Entry(self.input_frame)
        self.alphabet_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.states_label = ttk.Label(self.input_frame, text="States (comma-separated):")
        self.states_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.states_entry = ttk.Entry(self.input_frame)
        self.states_entry.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.initial_state_label = ttk.Label(self.input_frame, text="Initial State:")
        self.initial_state_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.initial_state_entry = ttk.Entry(self.input_frame)
        self.initial_state_entry.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.final_states_label = ttk.Label(self.input_frame, text="Final States (comma-separated):")
        self.final_states_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.final_states_entry = ttk.Entry(self.input_frame)
        self.final_states_entry.grid(row=3, column=1, padx=5, pady=5, sticky="ew")

        self.stack_alphabet_label = ttk.Label(self.input_frame, text="Stack Alphabet (comma-separated):")
        self.stack_alphabet_label.grid(row=4, column=0, padx=5, pady=5, sticky="w")
        self.stack_alphabet_entry = ttk.Entry(self.input_frame)
        self.stack_alphabet_entry.grid(row=4, column=1, padx=5, pady=5, sticky="ew")

        self.transitions_label = ttk.Label(self.input_frame, text="Transitions (one per line, format: q,a,S->p,PushString):")
        self.transitions_label.grid(row=5, column=0, padx=5, pady=5, sticky="nw")
        self.transitions_text = tk.Text(self.input_frame, height=10, width=50)
        self.transitions_text.grid(row=5, column=1, padx=5, pady=5, sticky="nsew")

        self.process_transitions_button = ttk.Button(self.input_frame, text="Process Transitions", command=self.process_transitions)
        self.process_transitions_button.grid(row=6, column=0, columnspan=2, padx=5, pady=5)


        # Input String Frame
        self.input_string_frame = ttk.LabelFrame(master, text="Simulation Input")
        self.input_string_frame.grid(row=0, column=1, padx=10, pady=10, sticky="nsew")

        self.input_string_label = ttk.Label(self.input_string_frame, text="Input String:")
        self.input_string_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.input_string_entry = ttk.Entry(self.input_string_frame)
        self.input_string_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.start_simulation_button = ttk.Button(self.input_string_frame, text="Start Simulation", command=self.start_simulation)
        self.start_simulation_button.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

        self.next_step_button = ttk.Button(self.input_string_frame, text="Next Step", command=self.step_simulation_gui, state=tk.DISABLED)
        self.next_step_button.grid(row=2, column=0, columnspan=2, padx=5, pady=5)


        # Output Frame
        self.output_frame = ttk.LabelFrame(master, text="Simulation Output")
        self.output_frame.grid(row=1, column=0, columnspan=2, padx=10, pady=10, sticky="nsew")

        self.current_state_label = ttk.Label(self.output_frame, text="Current State:")
        self.current_state_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.current_state_display = ttk.Label(self.output_frame, text="")
        self.current_state_display.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.stack_content_label = ttk.Label(self.output_frame, text="Stack Content:")
        self.stack_content_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.stack_content_display = ttk.Label(self.output_frame, text="")
        self.stack_content_display.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.remaining_input_label = ttk.Label(self.output_frame, text="Remaining Input:")
        self.remaining_input_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.remaining_input_display = ttk.Label(self.output_frame, text="")
        self.remaining_input_display.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.acceptance_status_label = ttk.Label(self.output_frame, text="Acceptance Status:")
        self.acceptance_status_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.acceptance_status_display = ttk.Label(self.output_frame, text="")
        self.acceptance_status_display.grid(row=3, column=1, padx=5, pady=5, sticky="ew")

        self.pda = None # Placeholder for the PDA instance
        self.parsed_transitions = {} # To store transitions after processing

    def process_transitions(self):
        transitions_input = self.transitions_text.get("1.0", "end-1c")
        transitions = {}
        errors = []

        for line_num, line in enumerate(transitions_input.splitlines()):
            line = line.strip()
            if not line:
                continue

            try:
                # Expected format: q,a,S->p,PushString
                parts = line.split("->")
                if len(parts) != 2:
                    errors.append(f"Line {line_num+1}: Invalid format. Expected 'q,a,S->p,PushString'.")
                    continue

                left_part = parts[0].split(",")
                right_part = parts[1].split(",")

                if len(left_part) != 3 or len(right_part) != 2:
                     errors.append(f"Line {line_num+1}: Invalid number of elements in transition rule.")
                     continue

                current_state, input_symbol, stack_top = left_part
                next_state, stack_push_string = right_part

                # Use 'epsilon' or a similar convention for epsilon transitions
                if input_symbol.lower() == 'epsilon':
                    input_symbol = ''
                if stack_top.lower() == 'epsilon':
                    stack_top = ''
                if stack_push_string.lower() == 'epsilon':
                    stack_push_string = ''


                # Store transitions in a dictionary: (current_state, input_symbol, stack_top) -> list of (next_state, stack_push_string)
                key = (current_state.strip(), input_symbol.strip(), stack_top.strip())
                value = (next_state.strip(), stack_push_string.strip())

                if key not in transitions:
                    transitions[key] = []
                transitions[key].append(value)

            except Exception as e:
                errors.append(f"Line {line_num+1}: Error parsing transition - {e}")


        if errors:
            messagebox.showerror("Transition Input Error", "\n".join(errors))
            self.parsed_transitions = {} # Clear transitions if invalid
        else:
            messagebox.showinfo("Transition Input", "Transitions processed successfully.")
            self.parsed_transitions = transitions
            print("Parsed Transitions:", self.parsed_transitions) # For debugging

    def start_simulation(self):
        # Get PDA definition from GUI entries
        states = [s.strip() for s in self.states_entry.get().split(',') if s.strip()]
        input_alphabet = [s.strip() for s in self.alphabet_entry.get().split(',') if s.strip()]
        stack_alphabet = [s.strip() for s in self.stack_alphabet_entry.get().split(',') if s.strip()]
        initial_state = self.initial_state_entry.get().strip()
        final_states = [s.strip() for s in self.final_states_entry.get().split(',') if s.strip()]
        input_string = self.input_string_entry.get().strip()

        if not self.parsed_transitions:
            messagebox.showwarning("Simulation Error", "Please process transitions first.")
            return

        # Create the PDA instance
        try:
            self.pda = PushdownAutomaton(
                states=states,
                input_alphabet=input_alphabet,
                stack_alphabet=stack_alphabet,
                initial_state=initial_state,
                final_states=final_states,
                transitions=self.parsed_transitions
            )
            self.pda.remaining_input = input_string # Set the full input string
            self.pda.input_index = 0 # Reset input index
            self.pda.reset() # Reset PDA state and stack

            self.update_gui_output()
            self.acceptance_status_display.config(text="") # Clear previous status
            self.next_step_button.config(state=tk.NORMAL) # Enable Next Step button
            self.start_simulation_button.config(state=tk.DISABLED) # Disable Start Simulation button

        except Exception as e:
            messagebox.showerror("PDA Creation Error", f"Error creating PDA: {e}")


    def step_simulation_gui(self):
        if self.pda is None:
            messagebox.showwarning("Simulation Error", "Please start a simulation first.")
            return

        status = self.pda.step_simulation()
        self.update_gui_output()

        if status == "halted":
            self.acceptance_status_display.config(text="Halted (Rejected)", foreground="red")
            self.next_step_button.config(state=tk.DISABLED) # Disable Next Step button
            self.start_simulation_button.config(state=tk.NORMAL) # Enable Start Simulation button
        elif status == "accepted":
             self.acceptance_status_display.config(text="Accepted", foreground="green")
             self.next_step_button.config(state=tk.DISABLED) # Disable Next Step button
             self.start_simulation_button.config(state=tk.NORMAL) # Enable Start Simulation button
        elif status == "continue":
            # Simulation continues, nothing to do here
            pass


    def update_gui_output(self):
        if self.pda:
            self.current_state_display.config(text=self.pda.current_state)
            # Display stack content from bottom to top
            self.stack_content_display.config(text="".join(self.pda.stack._items) if self.pda.stack._items else 'ε')
            self.remaining_input_display.config(text=self.pda.remaining_input[self.pda.input_index:] if self.pda.input_index < len(self.pda.remaining_input) else 'ε')
        else:
             self.current_state_display.config(text="N/A")
             self.stack_content_display.config(text="N/A")
             self.remaining_input_display.config(text="N/A")


# Create the main window
# root = tk.Tk()
# pda_gui = PDASimulatorGUI(root)
# root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
from tkinter.tix import Balloon # Import Balloon for tooltips

class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if not self.is_empty():
            return self._items.pop()
        return None

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        return None

    def is_empty(self):
        return not self._items

    def __str__(self):
        return "".join(self._items) # Display stack as a string

    def __repr__(self):
        return repr(self._items)

class PushdownAutomaton:
    def __init__(self, states, input_alphabet, stack_alphabet, initial_state, final_states, transitions):
        self.states = set(states)
        self.input_alphabet = set(input_alphabet)
        self.stack_alphabet = set(stack_alphabet)
        self.initial_state = initial_state
        self.final_states = set(final_states)
        self.transitions = transitions

        # Basic validation during PDA creation
        if not self.states:
            raise ValueError("States cannot be empty.")
        if not self.input_alphabet:
             raise ValueError("Input alphabet cannot be empty.")
        if not self.stack_alphabet:
             raise ValueError("Stack alphabet cannot be empty.")
        if self.initial_state not in self.states:
            raise ValueError(f"Initial state '{self.initial_state}' is not in the set of states.")
        if not self.final_states.issubset(self.states):
            raise ValueError("Final states must be a subset of the defined states.")
        # Add more detailed transition validation if needed


        self.current_state = initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0

        self.acceptance_condition = 'both' # Default

    def reset(self):
        self.current_state = self.initial_state
        self.stack = Stack()
        self.remaining_input = ""
        self.is_halted = False
        self.input_index = 0

    def get_transitions(self, current_state, input_symbol, stack_top):
        key = (current_state, input_symbol, stack_top)
        return self.transitions.get(key, [])

    def apply_transition(self, transition):
        next_state, stack_push_string = transition
        self.current_state = next_state

        # Pop the top of the stack if the transition rule required reading a stack symbol
        # (assuming get_transitions was called with a non-empty stack_top if applicable)
        if self.stack.peek() is not None: # Only pop if the stack is not empty
             self.stack.pop()


        # Push the stack_push_string characters onto the stack in reverse order
        # Handle empty string push (epsilon)
        if stack_push_string != '':
            for char in reversed(stack_push_string):
                self.stack.push(char)


    def step_simulation(self):
        if self.is_halted:
            if self.is_accepted():
                 return "accepted"
            else:
                 return "halted"

        current_state = self.current_state
        stack_top = self.stack.peek()
        current_symbol = self.remaining_input[self.input_index] if self.input_index < len(self.remaining_input) else ''

        # Try transitions with the actual input symbol
        possible_transitions = self.get_transitions(current_state, current_symbol, stack_top)

        # If no transitions with the input symbol, try epsilon transitions
        if not possible_transitions:
            possible_transitions = self.get_transitions(current_state, '', stack_top)
            input_consumed = False # Epsilon transition, input not consumed
        else:
            input_consumed = True # Non-epsilon transition, input is consumed


        if possible_transitions:
            # For simplicity, handle non-determinism by choosing the first valid transition
            chosen_transition = possible_transitions[0]
            self.apply_transition(chosen_transition)
            if input_consumed:
                self.input_index += 1 # Move to the next input symbol

            # After applying a transition, check if the simulation should halt
            # Simulation halts if no further transitions (input or epsilon) are possible
            next_symbol = self.remaining_input[self.input_index] if self.input_index < len(self.remaining_input) else ''
            if not self.get_transitions(self.current_state, next_symbol, self.stack.peek()) and \
               not self.get_transitions(self.current_state, '', self.stack.peek()):
                self.is_halted = True
                if self.is_accepted():
                    return "accepted"
                else:
                    return "halted"
            else:
                return "continue"

        else:
            # No valid transition found, halt simulation
            self.is_halted = True
            if self.is_accepted():
                 return "accepted"
            else:
                 return "halted"


    def is_accepted(self):
        """
        Checks if the input string is accepted by the PDA based on the defined acceptance condition.
        Current condition is final state AND empty stack after processing the entire input string.
        """
        # Acceptance by final state AND empty stack
        return self.input_index >= len(self.remaining_input) and self.stack.is_empty() and self.current_state in self.final_states


class PDASimulatorGUI:
    def __init__(self, master):
        self.master = master
        master.title("Pushdown Automaton Simulator")

        # Initialize tooltip manager
        self.balloon = Balloon(master)


        # Configure style
        style = ttk.Style()
        style.configure("TLabel", padding=5, font=('Arial', 10))
        style.configure("TButton", padding=5, font=('Arial', 10))
        style.configure("TEntry", padding=5, font=('Arial', 10))
        style.configure("TText", padding=5, font=('Arial', 10))
        style.configure("TLabelframe.Label", font=('Arial', 12, 'bold')) # Style for LabelFrame label


        # --- PDA Definition Frame ---
        self.input_frame = ttk.LabelFrame(master, text="PDA Definition")
        self.input_frame.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")
        # Configure column weights so entry fields expand
        self.input_frame.columnconfigure(1, weight=1)


        self.alphabet_label = ttk.Label(self.input_frame, text="Input Alphabet:")
        self.alphabet_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.alphabet_entry = ttk.Entry(self.input_frame)
        self.alphabet_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")
        self.balloon.bind(self.alphabet_entry, "Comma-separated list of input symbols (e.g., a,b)")

        self.states_label = ttk.Label(self.input_frame, text="States:")
        self.states_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.states_entry = ttk.Entry(self.input_frame)
        self.states_entry.grid(row=1, column=1, padx=5, pady=5, sticky="ew")
        self.balloon.bind(self.states_entry, "Comma-separated list of state names (e.g., q0,q1,q_accept)")

        self.initial_state_label = ttk.Label(self.input_frame, text="Initial State:")
        self.initial_state_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.initial_state_entry = ttk.Entry(self.input_frame)
        self.initial_state_entry.grid(row=2, column=1, padx=5, pady=5, sticky="ew")
        self.balloon.bind(self.initial_state_entry, "The name of the initial state (must be in States)")

        self.final_states_label = ttk.Label(self.input_frame, text="Final States:")
        self.final_states_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.final_states_entry = ttk.Entry(self.input_frame)
        self.final_states_entry.grid(row=3, column=1, padx=5, pady=5, sticky="ew")
        self.balloon.bind(self.final_states_entry, "Comma-separated list of final state names (must be in States)")

        self.stack_alphabet_label = ttk.Label(self.input_frame, text="Stack Alphabet:")
        self.stack_alphabet_label.grid(row=4, column=0, padx=5, pady=5, sticky="w")
        self.stack_alphabet_entry = ttk.Entry(self.input_frame)
        self.stack_alphabet_entry.grid(row=4, column=1, padx=5, pady=5, sticky="ew")
        self.balloon.bind(self.stack_alphabet_entry, "Comma-separated list of stack symbols (e.g., Z0,X,Y)")

        self.transitions_label = ttk.Label(self.input_frame, text="Transitions:")
        self.transitions_label.grid(row=5, column=0, padx=5, pady=5, sticky="nw")
        self.transitions_text = tk.Text(self.input_frame, height=8, width=40, font=('Arial', 10)) # Adjusted height/width
        self.transitions_text.grid(row=5, column=1, padx=5, pady=5, sticky="nsew")
        self.balloon.bind(self.transitions_text, "Enter transitions one per line. Format: q,a,S->p,PushString\n'epsilon' or empty string for epsilon.")

        self.process_transitions_button = ttk.Button(self.input_frame, text="Process Transitions", command=self.process_transitions)
        self.process_transitions_button.grid(row=6, column=0, columnspan=2, padx=5, pady=5)


        # --- Simulation Input Frame ---
        self.input_string_frame = ttk.LabelFrame(master, text="Simulation Input")
        self.input_string_frame.grid(row=0, column=1, padx=10, pady=10, sticky="nsew")
        self.input_string_frame.columnconfigure(1, weight=1) # Configure column weights


        self.input_string_label = ttk.Label(self.input_string_frame, text="Input String:")
        self.input_string_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.input_string_entry = ttk.Entry(self.input_string_frame)
        self.input_string_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")
        self.balloon.bind(self.input_string_entry, "The input string to simulate (symbols must be in Input Alphabet)")

        self.start_simulation_button = ttk.Button(self.input_string_frame, text="Start Simulation", command=self.start_simulation)
        self.start_simulation_button.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

        self.next_step_button = ttk.Button(self.input_string_frame, text="Next Step", command=self.step_simulation_gui, state=tk.DISABLED)
        self.next_step_button.grid(row=2, column=0, columnspan=2, padx=5, pady=5)

        self.reset_button = ttk.Button(self.input_string_frame, text="Reset Simulation", command=self.reset_gui)
        self.reset_button.grid(row=3, column=0, columnspan=2, padx=5, pady=5)


        # --- Simulation Output Frame ---
        self.output_frame = ttk.LabelFrame(master, text="Simulation Output")
        self.output_frame.grid(row=1, column=0, columnspan=2, padx=10, pady=10, sticky="nsew")
        self.output_frame.columnconfigure(1, weight=1) # Configure column weights


        self.current_state_label = ttk.Label(self.output_frame, text="Current State:")
        self.current_state_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.current_state_display = ttk.Label(self.output_frame, text="")
        self.current_state_display.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

        self.stack_content_label = ttk.Label(self.output_frame, text="Stack Content:")
        self.stack_content_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.stack_content_display = ttk.Label(self.output_frame, text="")
        self.stack_content_display.grid(row=1, column=1, padx=5, pady=5, sticky="ew")

        self.remaining_input_label = ttk.Label(self.output_frame, text="Remaining Input:")
        self.remaining_input_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.remaining_input_display = ttk.Label(self.output_frame, text="")
        self.remaining_input_display.grid(row=2, column=1, padx=5, pady=5, sticky="ew")

        self.acceptance_status_label = ttk.Label(self.output_frame, text="Acceptance Status:")
        self.acceptance_status_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.acceptance_status_display = ttk.Label(self.output_frame, text="")
        self.acceptance_status_display.grid(row=3, column=1, padx=5, pady=5, sticky="ew")


        self.pda = None
        self.parsed_transitions = {}

        # Configure master window grid weights for responsiveness
        master.columnconfigure(0, weight=1)
        master.columnconfigure(1, weight=1)
        master.rowconfigure(0, weight=1)
        master.rowconfigure(1, weight=1)


    def process_transitions(self):
        transitions_input = self.transitions_text.get("1.0", "end-1c")
        transitions = {}
        errors = []

        for line_num, line in enumerate(transitions_input.splitlines()):
            line = line.strip()
            if not line:
                continue

            try:
                # Expected format: q,a,S->p,PushString
                parts = line.split("->")
                if len(parts) != 2:
                    errors.append(f"Line {line_num+1}: Invalid format. Expected 'q,a,S->p,PushString'.")
                    continue

                left_part = parts[0].split(",")
                right_part = parts[1].split(",")

                if len(left_part) != 3 or len(right_part) != 2:
                     errors.append(f"Line {line_num+1}: Invalid number of elements in transition rule.")
                     continue

                current_state, input_symbol, stack_top = left_part
                next_state, stack_push_string = right_part

                # Use 'epsilon' or a similar convention for epsilon transitions
                if input_symbol.lower() == 'epsilon':
                    input_symbol = ''
                if stack_top.lower() == 'epsilon':
                    stack_top = ''
                if stack_push_string.lower() == 'epsilon':
                    stack_push_string = ''


                # Store transitions in a dictionary: (current_state, input_symbol, stack_top) -> list of (next_state, stack_push_string)
                key = (current_state.strip(), input_symbol.strip(), stack_top.strip())
                value = (next_state.strip(), stack_push_string.strip())

                if key not in transitions:
                    transitions[key] = []
                transitions[key].append(value)

            except Exception as e:
                errors.append(f"Line {line_num+1}: Error parsing transition - {e}")


        if errors:
            messagebox.showerror("Transition Input Error", "Errors found in transitions:\n" + "\n".join(errors))
            self.parsed_transitions = {} # Clear transitions if invalid
        else:
            messagebox.showinfo("Transition Input", "Transitions processed successfully.")
            self.parsed_transitions = transitions
            print("Parsed Transitions:", self.parsed_transitions) # For debugging

    def start_simulation(self):
        # Get PDA definition from GUI entries
        states_input = self.states_entry.get().strip()
        alphabet_input = self.alphabet_entry.get().strip()
        stack_alphabet_input = self.stack_alphabet_entry.get().strip()
        initial_state = self.initial_state_entry.get().strip()
        final_states_input = self.final_states_entry.get().strip()
        input_string = self.input_string_entry.get().strip()

        states = [s.strip() for s in states_input.split(',') if s.strip()] if states_input else []
        input_alphabet = [s.strip() for s in alphabet_input.split(',') if s.strip()] if alphabet_input else []
        stack_alphabet = [s.strip() for s in stack_alphabet_input.split(',') if s.strip()] if stack_alphabet_input else []
        final_states = [s.strip() for s in final_states_input.split(',') if s.strip()] if final_states_input else []


        if not self.parsed_transitions:
            messagebox.showwarning("Simulation Error", "Please process transitions first.")
            return

        # Create the PDA instance with input validation
        try:
            self.pda = PushdownAutomaton(
                states=states,
                input_alphabet=input_alphabet,
                stack_alphabet=stack_alphabet,
                initial_state=initial_state,
                final_states=final_states,
                transitions=self.parsed_transitions
            )
            self.pda.remaining_input = input_string # Set the full input string
            self.pda.input_index = 0 # Reset input index
            self.pda.reset() # Reset PDA state and stack

            self.update_gui_output()
            self.acceptance_status_display.config(text="", foreground="black") # Clear previous status and color
            self.next_step_button.config(state=tk.NORMAL) # Enable Next Step button
            self.start_simulation_button.config(state=tk.DISABLED) # Disable Start Simulation button

        except ValueError as e:
             messagebox.showerror("PDA Definition Error", f"Invalid PDA definition: {e}")
             self.pda = None # Clear PDA if creation failed
        except Exception as e:
            messagebox.showerror("PDA Creation Error", f"An unexpected error occurred creating PDA: {e}")
            self.pda = None # Clear PDA if creation failed


    def step_simulation_gui(self):
        if self.pda is None:
            messagebox.showwarning("Simulation Error", "Please start a simulation first.")
            return

        status = self.pda.step_simulation()
        self.update_gui_output()

        if status == "halted":
            self.acceptance_status_display.config(text="Halted (Rejected)", foreground="red")
            self.next_step_button.config(state=tk.DISABLED)
            self.start_simulation_button.config(state=tk.NORMAL)
        elif status == "accepted":
             self.acceptance_status_display.config(text="Accepted", foreground="green")
             self.next_step_button.config(state=tk.DISABLED)
             self.start_simulation_button.config(state=tk.NORMAL)
        elif status == "continue":
            pass # Simulation continues

    def update_gui_output(self):
        if self.pda:
            self.current_state_display.config(text=self.pda.current_state)
            self.stack_content_display.config(text=str(self.pda.stack) if not self.pda.stack.is_empty() else 'ε')
            remaining = self.pda.remaining_input[self.pda.input_index:]
            self.remaining_input_display.config(text=remaining if remaining else 'ε')
        else:
             self.current_state_display.config(text="N/A")
             self.stack_content_display.config(text="N/A")
             self.remaining_input_display.config(text="N/A")
             self.acceptance_status_display.config(text="N/A", foreground="black") # Reset status display on full reset


    def reset_gui(self):
        """Clears all input fields and output displays, and resets the PDA."""
        self.alphabet_entry.delete(0, tk.END)
        self.states_entry.delete(0, tk.END)
        self.initial_state_entry.delete(0, tk.END)
        self.final_states_entry.delete(0, tk.END)
        self.stack_alphabet_entry.delete(0, tk.END)
        self.transitions_text.delete("1.0", tk.END)
        self.input_string_entry.delete(0, tk.END)

        self.pda = None # Clear PDA instance
        self.parsed_transitions = {} # Clear parsed transitions

        self.update_gui_output() # Reset output displays

        self.next_step_button.config(state=tk.DISABLED)
        self.start_simulation_button.config(state=tk.NORMAL)
        self.acceptance_status_display.config(text="", foreground="black") # Clear acceptance status


# Create the main window
root = tk.Tk()
pda_gui = PDASimulatorGUI(root)
root.mainloop()
```
